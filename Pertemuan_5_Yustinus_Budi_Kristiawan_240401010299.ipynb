```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df = sns.load_dataset('tips')

print(df.head())
print('\nJumlah baris dan kolom:', df.shape)
print('\nTipe data:')
print(df.dtypes)
print('\nStatistik deskriptif:')
print(df.describe().round(2))
```

# **Membuat Dashboard Visualisasi Statis**

```python
fig = plt.figure(figsize=(18, 12))
fig.suptitle('Dashboard Analisis Dataset Tips', fontsize=20, fontweight='bold', y=0.98)

gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.35, wspace=0.30)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# GRAFIK 1 – BAR CHART

avg_total_bill = df.groupby('day')['total_bill'].mean()

avg_total_bill.plot(
    kind='bar',
    ax=ax1,
    color='skyblue',
    edgecolor='black'
)

ax1.set_title('Rata-rata Total Tagihan per Hari')
ax1.set_xlabel('Hari')
ax1.set_ylabel('Rata-rata Total Bill (USD)')

# GRAFIK 2 – HISTOGRAM

sns.histplot(
    data=df,
    x='total_bill',
    kde=True,
    color='orange',
    ax=ax2
)

ax2.set_title('Distribusi Total Tagihan')
ax2.set_xlabel('Total Bill (USD)')
ax2.set_ylabel('Frekuensi')

# GRAFIK 3 – BOXPLOT

sns.boxplot(
    data=df,
    x='day',
    y='tip',
    hue='sex',
    palette='Set2',
    ax=ax3
)

ax3.set_title('Distribusi Tip Berdasarkan Hari dan Gender')
ax3.set_xlabel('Hari')
ax3.set_ylabel('Tip (USD)')

# GRAFIK 4 – SCATTER PLOT

sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    sizes=(30, 200),
    palette='Set1',
    alpha=0.7,
    ax=ax4
)

sns.regplot(
    data=df,
    x='total_bill',
    y='tip',
    scatter=False,
    color='black',
    ax=ax4
)
plt.show()
```

# **Analisis dan Narasi Dashboard**

# **Grafik 1 – Bar Chart: Rata-rata Total Tagihan per Hari**

**What?**

Grafik menunjukkan rata-rata total tagihan paling tinggi ada di hari Sabtu dan Minggu. Sementara itu, hari Kamis memiliki rata-rata tagihan yang lebih rendah.

**So What?**

Ini menunjukkan bahwa pelanggan lebih banyak makan atau belanja saat akhir pekan dibanding hari biasa.

**Now What?**

Restoran bisa menambah stok makanan dan jumlah pegawai saat akhir pekan karena pelanggan biasanya lebih ramai.

# **Grafik 2 – Histogram: Distribusi Total Tagihan**

**What?**

Sebagian besar total tagihan pelanggan berada di kisaran 10–20 USD. Hanya sedikit pelanggan yang memiliki tagihan sangat besar.

**So What?**

Artinya, kebanyakan pelanggan melakukan transaksi dalam jumlah sedang.

**Now What?**

Restoran bisa membuat promo untuk pelanggan agar jumlah pembelian mereka meningkat.

# **Grafik 3 – Boxplot: Distribusi Tip Berdasarkan Hari dan Gender**

**What?**

Jumlah tip pada hari Sabtu dan Minggu terlihat lebih bervariasi dibanding hari lainnya. Ada juga beberapa pelanggan yang memberi tip cukup besar.

**So What?**

Hal ini menunjukkan bahwa perilaku pelanggan saat memberi tip berbeda-beda, terutama saat akhir pekan.

**Now What?**

Restoran bisa meningkatkan kualitas pelayanan saat akhir pekan agar pelanggan lebih puas dan memberi tip lebih baik.

# **Grafik 4 – Scatter Plot: Hubungan Total Bill dan Tip**

**What?**

Grafik menunjukkan semakin besar total tagihan, biasanya tip yang diberikan juga semakin besar.

**So What?**

Ini berarti pelanggan yang membeli lebih banyak cenderung memberi tip lebih tinggi.

**Now What?**

Restoran bisa menawarkan paket menu atau promo keluarga untuk meningkatkan total pembelian pelanggan.
