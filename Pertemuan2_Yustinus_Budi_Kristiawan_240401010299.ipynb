```python
import pandas as pd
import numpy as np
```

Melakukan import library eksternal

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

Data berhasil dimuat untuk menganalisa struktur data set

**Q1 Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?**

```python
print(df.shape)  # Menampilkan (jumlah_baris, jumlah_kolom)
print(df.columns) # Menampilkan daftar nama kolom
```

**Q2 Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)
per kolom?**

```python
missing_data = df.isnull().sum() # Menampilkan data yang kosong

percentage = (missing_data / len(df)) * 100
print(pd.DataFrame({'Jumlah': missing_data, 'Persentase (%)': percentage})) # Mencari hasil presentasi dari data yang kosong
```

**Q3 Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy
atau Pandas.**

```python
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase selamat: {persen_selamat:.2f}%")
```

**Q4 Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-
rata usianya?**

```python
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print(f"Jumlah: {len(wanita_kelas1)}")
print(f"Rata-rata usia: {wanita_kelas1['Age'].mean():.2f}")
```

**Q5 Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas
penumpang. Kelas mana yang paling tinggi?**

```python
status_kelas = df.groupby('Pclass')['Survived'].mean()
print(status_kelas)
```

**Q6 Apakah orang yang sendirian lebih mudah menyelamatkan diri, atau mereka yang berkeluarga?**

```python
# Menggabungkan SibSp (saudara/pasangan) dan Parch (orang tua/anak)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Melihat tingkat keselamatan berdasarkan ukuran keluarga
print(df.groupby('FamilySize')['Survived'].mean())
```
